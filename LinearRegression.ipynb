```python
import numpy as np
import matplotlib.pyplot as plt

# Example dataset
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([20, 35, 50, 60, 75, 90])

# Step 1: Initialize parameters
beta_0 = 0.0  # intercept
beta_1 = 0.0  # slope

# Hyperparameters
learning_rate = 0.01
epochs = 1000
n = len(x)

# Step 2: Gradient Descent Loop
for _ in range(epochs):
    y_pred = beta_0 + beta_1 * x  # predictions

    # Compute gradients
    d_beta_0 = (-2/n) * np.sum(y - y_pred)
    d_beta_1 = (-2/n) * np.sum(x * (y - y_pred))

    # Update parameters
    beta_0 -= learning_rate * d_beta_0
    beta_1 -= learning_rate * d_beta_1

print(f"Learned Slope (β1): {beta_1:.2f}")
print(f"Learned Intercept (β0): {beta_0:.2f}")

# Step 3: Predictions with learned parameters
y_pred = beta_0 + beta_1 * x

# Step 4: Visualization
plt.scatter(x, y, color="blue", label="Actual data")
plt.plot(x, y_pred, color="red", linewidth=2, label="Regression line")
plt.xlabel("Hours studied")
plt.ylabel("Exam score")
plt.legend()
plt.show()
```
